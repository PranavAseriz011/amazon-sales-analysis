```python
# importing the pandas and read the csv file 

import pandas as pd

df = pd.read_csv("Amazon sale Report.csv")

df.head()
```

```python
#Print all column names 

print(df.columns)
```

```python
#printing the all info of the data in the csv file 

df.info()
```

```python
#checking the null values of the data

df.isnull().sum()
```

```python
#checking duplicates entry also 

df.duplicated().sum()
```

```python
#cheking the count of the values 

df["Category"].value_counts() 
```

```python
#grouping the category and sales cahnnel and sum them to print category sales 

Category_sales = df.groupby("Category")["Sales Channel "].sum()
print(Category_sales)
```

```python
#groupby the category and amount sum 

Category_sales = df.groupby("Category")["Amount"].sum()
```

```python
print(Category_sales)
```

```python
#ploting the bargraph in matplotlib category vs sales amount

import matplotlib.pyplot as plt

Category_sales.plot(kind="bar")

plt.title("Total sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales Amount")

plt.show()
```

```python
#grouping the ship-state and amount

state_sales = df.groupby("ship-state")["Amount"].sum()

state_sales.sort_values(ascending=False).head(10)
```

```python
#ploting the graph of category vs qty

qty_sales = df.groupby("Category")["Qty"].sum()

qty_sales.plot(kind="bar")
plt.show()
```

```python
# Top 10 states by sales

state_sales = df.groupby("ship-state")["Amount"].sum()

top_states = state_sales.sort_values(ascending=False).head(10)

top_states.plot(kind="bar")

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales Amount")

plt.show()
```

```python
#Most ordered categories

qty_sales = df.groupby("Category")["Qty"].sum()

qty_sales.plot(kind="bar")

plt.title("Quantity Sold by Category")
plt.xlabel("Category")
plt.ylabel("Quantity")

plt.show()
```

```python
#order status analysis

df["Status"].value_counts().plot(kind="bar")

plt.title("Order Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")

plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df["Month"] = df["Date"].dt.month
```

```python
#monthly sales analysis 

monthly_sales = df.groupby("Month")["Amount"].sum()

print(monthly_sales)
```

```python
monthly_sales.plot(kind="line")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```
